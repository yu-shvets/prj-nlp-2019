#### 2. Перевірте роботу [SnowballStem](http://snowballstem.org/) для спільнокореневих слів. Напишіть ваші спостереження.

1. truth, truthful, truthfulness, countertruth, untruthful, truthology
2. flaw, flaws, flawed, flawless, flawlessness, flawlessly, 
3. лес, лесной, лесник, лесничий, лесничество, пролесье
4. окно, окошко, подоконник, оконный, окнище

````python
import snowballstemmer
en_stemmer = snowballstemmer.stemmer('english')
ru_stemmer = snowballstemmer.stemmer('russian')

list1 = 'truth, truthful, truthfulness, countertruth, untruthful, truthology'.split(', ')
list2 = 'flaw, flaws, flawed, flawless, flawlessness, flawlessly'.split(', ')
list3 = 'лес, лесной, лесник, лесничий, лесничество, пролесье'.split(', ')
list4 = 'окно, окошко, подоконник, оконный, окнище'.split(', ')
````

````python
for i in list1:
    print(en_stemmer.stemWord(i))
````

Перші три слова застемились до одного truth, бо вони мають один корінь. Аналогічно і інші, стеми утворюються від того, що забирається словотворчі афікси.

````python
for i in list2:
    print(en_stemmer.stemWord(i))
````

Суфікс less залишається після процедури стемінгу, що є логічним, оскільки цей суфікс сильно змінює значення слова.

````python
for i in list3:
    print(ru_stemmer.stemWord(i))
````

````python
for i in list4:
    print(ru_stemmer.stemWord(i))
````

Даний стемінг обрізав такі останні буквосполучення: -ой, -ий, -о, -е, -ный, -е. Що наводить мене на думку, що він побудований на правилах.
