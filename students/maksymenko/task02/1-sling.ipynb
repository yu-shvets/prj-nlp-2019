````
import Stemmer
# help(Stemmer)

en_stemmer = Stemmer.Stemmer('english')
ru_stemmer = Stemmer.Stemmer('russian')

print(en_stemmer.stemWords(['truth', 'truthful', 'truthfulness', 'countertruth', 'untruthful', 'truthology']))
print(en_stemmer.stemWords(['flaw', 'flaws', 'flawed', 'flawless', 'flawlessness', 'flawlessly']))
print(ru_stemmer.stemWords(['лес', 'лесной', 'лесник', 'лесничий', 'лесничество', 'пролесье']))
print(ru_stemmer.stemWords(['окно', 'окошко', 'подоконник', 'оконный', 'окнище']))
````

У випадку з `truth, trughful, trughfulness`  стеммер приводить все до кореня `trurh`, тоді як у словах `countertrurh, untrughful, truthology` видіяє нові сутності.  Також розрізняє сутності `flaw` та `flawless` і не приводить все до `flaw`.
Щодо россійської мови - тут стеммер працює гірше, просто відкидає закінчення, хоча мав би проводити `лесник` та `лесничий` до одного кореня. Хоча для слів `пожарник` та `пожарный` цього не слід було б робити.

````
from nltk.stem.snowball import SnowballStemmer

# print(" ".join(SnowballStemmer.languages))

nltk_snowball_stemmer_en = SnowballStemmer('english')
nltk_snowball_stemmer_ru = SnowballStemmer('russian')


truth = ['truth', 'truthful', 'truthfulness', 'countertruth', 'untruthful', 'truthology']
flaw = ['flaw', 'flaws', 'flawed', 'flawless', 'flawlessness', 'flawlessly']
les = ['лес', 'лесной', 'лесник', 'лесничий', 'лесничество', 'пролесье']
okno = ['окно', 'окошко', 'подоконник', 'оконный', 'окнище']

def stem_words(stemmer, list):
  stems = [nltk_snowball_stemmer_en.stem(word) for word in list]
  print(stems)
    
stem_words(nltk_snowball_stemmer_en, truth)
stem_words(nltk_snowball_stemmer_en, flaw)
stem_words(nltk_snowball_stemmer_ru, les)
stem_words(nltk_snowball_stemmer_ru, okno)

````

nltk стеммер для англійскої мови працює ідентично пайстеммеру, а от для россійської взагалі не працює, або я не вмію його готувати.
