# Робота з даними 

```python
data.dropna(inplace=True)
```

```python
data = data[data['Rating'] != 0]
```

```python
target = ['Rating']
feature_names = ['Comment', 'Advantage', 'Disadvantage']

data_train, data_test, y_train, y_test = train_test_split(data[feature_names], data[target], 
                                                          test_size=0.3, stratify=data[target])
```

```python
def process_text(text):
    text = re.sub(r'\W+', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.lower()
    return text
```

```python
data_train['Comment'] = data_train['Comment'].apply(lambda x: process_text(x))
data_train['Advantage'] = data_train['Advantage'].apply(lambda x: process_text(x))
data_train['Disadvantage'] = data_train['Disadvantage'].apply(lambda x: process_text(x))
```

```python
data_test['Comment'] = data_test['Comment'].apply(lambda x: process_text(x))
data_test['Advantage'] = data_test['Advantage'].apply(lambda x: process_text(x))
data_test['Disadvantage'] = data_test['Disadvantage'].apply(lambda x: process_text(x))
```

```python
count_vect = CountVectorizer()
nb = MultinomialNB()
```

```python
X_train = count_vect.fit_transform(data_train['Comment'])
X_test = count_vect.transform(data_test['Comment'])
```

```python
nb.fit(X_train, y_train)
```

## Отримані результати 

```python
report = classification_report(y_test, nb.predict(X_test))
print(report)
```
