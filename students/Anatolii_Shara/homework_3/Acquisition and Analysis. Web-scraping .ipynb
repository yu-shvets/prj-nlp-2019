```python
import bs4 as bs
import urllib.request 
import requests
from lxml import html
from urllib.request import urlopen

sauce = urllib.request.urlopen('http://forum.lvivport.com/threads/chi-varto-jti-na-kursi.112092/').read()
 
soup = bs.BeautifulSoup(sauce,'lxml')
print(soup)
```

```python
print (soup.find_all('div'))
```

```python
for paragraph in soup.find_all('div'):
    print(paragraph.text)
```

```python
sauce = urllib.request.urlopen('http://forum.lvivport.com/forums/divochi-posidenki.194/').read()
 
soup = bs.BeautifulSoup(sauce,'lxml')

nav = soup.nav()

body = soup.body
for paragraph in body.find_all('p'):
    print (paragraph.text)
    
for div in soup.find_all('div', class_='body'):
    print(div.text)
```
