## Task

Write a query to collect all relations from dbpedia for every individual person listed in it. The task requires running a SPARQL request at <https://dbpedia.org/sparql>.

 ## Solution

```python
from SPARQLWrapper import SPARQLWrapper, JSON
    
sparql = SPARQLWrapper("http://dbpedia.org/sparql")
sparql.setQuery(
    """
    PREFIX dbo: <http://dbpedia.org/ontology/>
    
    SELECT distinct ?subj, ?predicate, ?obj WHERE 
    { 
        ?subj a dbo:Person . 
        ?subj ?predicate ?obj .
    } LIMIT 100
    """
)
    
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
                    
for result in results['results']['bindings']:
    print(result)
```
