### Wiktionary synonyms finder

I'm using Wiktionary dump for Norwegian language: https://dumps.wikimedia.org/nowiktionary/20190301/nowiktionary-20190301-pages-articles-multistream.xml.bz2

```python
import re
import xml.sax
import bz2
```

```python
class SynonymsFinder(xml.sax.ContentHandler):
    def __init__(self):
        self.tag = ''
        self.ns = ''
        self.word = ''
        self.lang = ''
        self.sect = ''
        self.syns = []
        self.re_lang = re.compile(r'(?<!===)(?<===)\w+')
        self.re_sect = re.compile(r'(?<=====)\w+')
        self.re_syns = re.compile(r'(?<=\[\[)[^\]]+(?=\]\])')
    
    def startElement(self, tag, attrs):
        self.tag = tag
        
    def endElement(self, tag):
        if self.tag == 'text' and self.ns == '0' and self.syns:
            print(f'\033[1;34m{self.word}\033[0m: {", ".join(self.syns)}')
        self.tag = ''
        
    def characters(self, content):
        if self.tag == 'ns': self.ns = content.strip()
        if self.tag == 'title':
            self.word = content.strip()
            self.lang = ''
            self.sect = ''
            self.syns = []
        
        if self.tag == 'text' and self.ns == '0':
            match = self.re_lang.search(content)
            if match: self.lang = match.group(0)
            
            match = self.re_sect.search(content)
            if match: self.sect = match.group(0)
                
            if self.lang == 'Norsk' and self.sect == 'Synonymer':
                match = self.re_syns.findall(content)
                self.syns.extend([
                    w.split('|')[1] if '|' in w else w for w in match
                ])
```

```python
parser = xml.sax.make_parser()
parser.setContentHandler( SynonymsFinder() )
parser.parse(bz2.open('nowiktionary-20190301-pages-articles-multistream.xml.bz2'))
```
