–®—É–∫–∞—î–º–æ –ø—Ä–∏–∫–ª–∞–¥–∏ –¥–ª—è –∑–∞–¥–∞–Ω–∏—Ö **—Ç–∏–ø—ñ–≤** –∑–∞–ª–µ–∂–Ω–æ—Å—Ç–µ–π.

```python
from collections import Counter
import spacy

nlp = spacy.load('en_core_web_lg')
```

```python
example_masks = ['VERB	prep	ADP',
                 'NOUN	prep	ADP',
                 'ADV	advmod	PART',
                 'ADP	pobj	NOUN',
                 'NOUN	det	DET',
                 'VERB	nsubj	NOUN',
                 'NOUN	compound	NOUN',
                 'NOUN	amod	ADJ',
                 'VERB	dobj	NOUN',
                 'VERB	aux	VERB',
                 'VERB	aux	PART',
                 'VERB	advmod	ADV',
                 'ADV	advmod	ADV',
                 'VERB	cc	CCONJ',
                 'NOUN	cc	CCONJ',
                 'VERB	ccomp	VERB',
                 'NOUN	poss	ADJ',
                 'VERB	mark	ADP',
                 'VERB	advcl	VERB',
                 'VERB	xcomp	VERB',
                 'NOUN	relcl	VERB',
                 'VERB	attr	NOUN',
                 'VERB	neg	ADV',
                 'NOUN	nummod	NUM',
                 'NOUN	appos	PROPN',
                 'PROPN	appos	NOUN',
                 'VERB	acomp	ADJ',
                 'PROPN	case	PART',
                 'VERB	npadvmod	NOUN',
                 'NOUN	acl	VERB',
                 'ADP	pcomp	VERB',
                 'VERB	prt	PART',
                 'VERB	nsubjpass	NOUN',
                 'NOUN	nmod	PROPN',
                 'VERB	agent	ADP',
                 'VERB	expl	ADV',
                 'VERB	intj	INTJ',
                 'VERB	csubj	VERB',
                 'VERB	auxpass	VERB',
                 'NUM	quantmod	ADP',
                 'VERB	dative	ADP',
                 'VERB	oprd	ADJ',
                 'VERB	dep	NOUN'
                ]

found_examples: int = 0
example_index = Counter()
```

```python
with open('blog2008.txt') as fo:
    lines = fo.readlines()
```

#### –ü—Ä–∏–∫–ª–∞–¥–∏

```python
for i, line in enumerate(lines):
    if len(line) < 30 or len(line) > 70 or i in [11,19,140, 480,504, 557,782,785,1390,7466]:
        continue

    printed: bool = False

    doc = nlp(line)
    for token in doc:
        key = f'{token.head.pos_}\t{token.dep_}\t{token.pos_}'
        if key in example_masks:
            if not example_index[key]:
                if not printed:
                    print(f'\n{i}üîπ{line.strip()}')
                    printed = True
                example_index[key] = True
                found_examples += 1
                print(f'\u001b[34;1m{key}\u001b[0m\t{token.head.text} ‚Üí {token.text}')
            else:
                if printed:
                    print(f'\u001b[34m{key}\u001b[0m\t{token.head.text} ‚Üí {token.text}')

    if found_examples == len(example_masks):
        break
```
